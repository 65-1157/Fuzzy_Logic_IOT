## Triangular_parâmetros_especialistas

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

# variáveis de entrada ou antecedentes, que influenciam a avaliação do serviço
Availability = ctrl.Antecedent(np.arange(0.1, 101.1, .01), 'availability')
Reliability = ctrl.Antecedent(np.arange(5.9,98.7, .5), 'reliability')
Response_Time = ctrl.Antecedent(np.arange(45,30782, 1), 'response_time')
Lattency = ctrl.Antecedent(np.arange(31.5,30782, .5), 'lattency')
Throughput = ctrl.Antecedent(np.arange(0.1,30, .1), 'throughput')

# variável de saída ou consequente, avaliação do serviço
WsRF = ctrl.Consequent(np.arange(30, 101, .25), 'WsRF')

# DEFINIÇÕES DAS FUNÇÕES DE PERTINÊNCIA -- sobre todas as variáveis: entrada & saída
# Availability
Availability['baixa'] = fuzz.trimf(Availability.universe, [0, 20, 35])
Availability['media'] = fuzz.trimf(Availability.universe, [15,50,65])
Availability['alta'] = fuzz.trimf(Availability.universe, [45,70,101])

# Reliability
Reliability['baixo'] = fuzz.trimf(Reliability.universe, [5,15,30])
Reliability['medio'] = fuzz.trimf(Reliability.universe, [20,50,75])
Reliability['alto'] = fuzz.trimf(Reliability.universe, [60,80,99])

Response_Time['baixo'] = fuzz.trimf(Response_Time.universe, [45,10000,12000])
Response_Time['medio'] = fuzz.trimf(Response_Time.universe, [8000,20000,27000])
Response_Time['alto'] = fuzz.trimf(Response_Time.universe, [24000,28000,30782])

Lattency['baixo'] = fuzz.trimf(Lattency.universe, [31,9000,12000])
Lattency['medio'] = fuzz.trimf(Lattency.universe, [8000,20000,27000])
Lattency['alto'] = fuzz.trimf(Lattency.universe, [21000,27000,30782])

Throughput['baixo'] = fuzz.trimf(Throughput.universe,[0.1,0.1,7.5])
Throughput['medio'] = fuzz.trimf(Throughput.universe,[5,15,22.5])
Throughput['alto'] = fuzz.trimf(Throughput.universe,[15,30,30])

# WsRF
WsRF['ruim'] = fuzz.trimf(WsRF.universe, [30,30,50])
WsRF['medio'] = fuzz.trimf(WsRF.universe, [45,60,80])
WsRF['bom'] = fuzz.trimf(WsRF.universe, [70,90,101])

# VISUALIZAÇÃO DAS FUNÇÕES DE PERTINÊNCIA

# regra 1 - 
regra1 = ctrl.Rule(Availability['baixa'] & Reliability['baixo'] & Response_Time['baixo'] & Lattency['baixo']
 & Throughput['baixo'], WsRF['ruim'])

# regra 2 - 
regra2 = ctrl.Rule(Availability['media'] & Reliability['medio'] & Response_Time['medio'] & Lattency['medio'] 
 & Throughput['medio'], WsRF['medio'])

# regra 3 - 
regra3 = ctrl.Rule(Availability['alta'] & Reliability['alto'] & Response_Time['alto'] & Lattency['alto'] 
 & Throughput['alto'] , WsRF['bom'])
# com 3 regras não é possível calcular: Total area is zero in defuzzification!
# acrescer regras para tornar cálculo possível

# regra 4 (adicional)
regra4 = ctrl.Rule(Availability['alta'] | Reliability['medio'] & Response_Time['medio'] & Throughput['alto'] &
        Lattency['medio'], WsRF['medio'])

# regra 5 (adicional)
regra5 = ctrl.Rule(Availability['media'] | Reliability['medio'] & Response_Time['alto'] & Throughput['medio'] &
                   Lattency['alto'], WsRF['ruim'])

regra_ctrl = ctrl.ControlSystem([regra1, regra2, regra3, regra4, regra5])
engine = ctrl.ControlSystemSimulation(regra_ctrl)
```

```python
# passa as predições dos modelos para suas respectivas variáveis de entrada
z =list(range(0,365))
s = 0
for i in z:
    engine.input['availability'] = av[i]
    engine.input['reliability'] = rb[i]
    engine.input['response_time'] = rt[i]
    engine.input['lattency'] = lt[i]
    engine.input['throughput'] = thr[i]
    engine.compute()
    e = (ws[i]-engine.output['WsRF'])/ws[i]
    ep = e * 100
    s = s + ep
    ec = s/(i+1)
    print(i)
    print(engine.output['WsRF'])
    print(ep)
    print(ec)
```

## Gaussiano_parâmetros_especialistas

```python
Availability = ctrl.Antecedent(np.arange(0.1, 101.1, .01), 'availability')
Reliability = ctrl.Antecedent(np.arange(5.9,98.7, .5), 'reliability')
Response_Time = ctrl.Antecedent(np.arange(45,30782, 1), 'response_time')
Lattency = ctrl.Antecedent(np.arange(31.5,30782, .5), 'lattency')
Throughput = ctrl.Antecedent(np.arange(0.1,30, .1), 'throughput')
# DEFINIÇÕES DAS FUNÇÕES DE PERTINÊNCIA -- sobre todas as variáveis: entrada & saída
# Availability
Availability['baixa'] = fuzz.gaussmf(Availability.universe,20,5)
Availability['media'] = fuzz.gaussmf(Availability.universe,50,10)
Availability['alta'] = fuzz.gaussmf(Availability.universe,70,10)

# Reliability
Reliability['baixo'] = fuzz.gaussmf(Reliability.universe,15,10)
Reliability['medio'] = fuzz.gaussmf(Reliability.universe,50,10)
Reliability['alto'] = fuzz.gaussmf(Reliability.universe,80,10)

Response_Time['baixo'] = fuzz.gaussmf(Response_Time.universe,10000,1000)
Response_Time['medio'] = fuzz.gaussmf(Response_Time.universe,16000,2000)
Response_Time['alto'] = fuzz.gaussmf(Response_Time.universe, 28000,2000)

Throughput['baixo'] = fuzz.gaussmf(Throughput.universe,4,1)
Throughput['medio'] = fuzz.gaussmf(Throughput.universe,13,3)
Throughput['alto'] = fuzz.gaussmf(Throughput.universe, 23,1)

Lattency['baixo'] = fuzz.gaussmf(Lattency.universe,9000,1500)
Lattency['medio'] = fuzz.gaussmf(Lattency.universe,20000,2000)
Lattency['alto'] = fuzz.gaussmf(Lattency.universe,27000,1500)

# WsRF
WsRF['ruim'] = fuzz.gaussmf(WsRF.universe,40,5)
WsRF['medio'] = fuzz.gaussmf(WsRF.universe, 65,5)
WsRF['bom'] = fuzz.gaussmf(WsRF.universe, 85,5)
```

```python
# regra 1 - 
regra1 = ctrl.Rule(Availability['baixa'] & Reliability['baixo'] & Response_Time['alto'] & Lattency['alto']
 & Throughput['baixo'], WsRF['ruim'])

# regra 2 - 
regra2 = ctrl.Rule(Availability['media'] & Reliability['medio'] & Response_Time['medio'] & Lattency['medio'] 
 & Throughput['medio'], WsRF['medio'])

# regra 3 - 
regra3 = ctrl.Rule(Availability['alta'] & Reliability['alto'] & Response_Time['baixo'] & Lattency['baixo'] 
 & Throughput['alto'] , WsRF['bom'])
# com 3 regras, o gaussiano resolve o que o triangular não consegue.
# acrescentando novas regras para diminuir o erro de 91,9%, mantendo gaussiano.

# regra 4 (adicional)
regra4 = ctrl.Rule(Availability['alta'] | Reliability['medio'] & Response_Time['medio'] & Throughput['alto'] &
        Lattency['medio'], WsRF['medio'])

# regra 5 (adicional)
regra5 = ctrl.Rule(Availability['media'] | Reliability['medio'] & Response_Time['alto'] & Throughput['medio'] &
                   Lattency['alto'], WsRF['ruim'])

regra_ctrl = ctrl.ControlSystem([regra1, regra2, regra3, regra4, regra5])
engine = ctrl.ControlSystemSimulation(regra_ctrl)

# com as 2 regras adicionais, o erro cai para 8%, mantendo gaussiano.
```

```python
# passa as predições dos modelos para suas respectivas variáveis de entrada
z =list(range(0,365))
s = 0
for i in z:
    engine.input['availability'] = av[i]
    engine.input['reliability'] = rb[i]
    engine.input['response_time'] = rt[i]
    engine.input['lattency'] = lt[i]
    engine.input['throughput'] = thr[i]
    engine.compute()
    e = (ws[i]-engine.output['WsRF'])/ws[i]
    ep = e * 100
    s = s + ep
    ec = s/(i+1)
    print(i)
    print(engine.output['WsRF'])
    print(ep)
    print(ec)
```
